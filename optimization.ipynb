```python
# Import the packages we need

import cvxpy as cvx
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# Create the dataset

m_true = 0.5
b_true = 0.5

x_train = np.linspace(0, 3, num=40)
y_train = m_true*x_train + b_true + 0.5*np.random.random((len(x_train)))

plt.plot(x_train, y_train, '.')
```

```python
# Formulate the optimization problem

m = cvx.Variable()
b = cvx.Variable()

y_pred = m*x_train + b

loss = cvx.norm(y_pred - y_train)
prob = cvx.Problem(cvx.Minimize(loss))
```

```python
# Solve the problem

prob.solve()
print("m:", m.value, "b:", b.value)
```

```python
# Plot the solution

x_test = np.linspace(0, 3, num=20)
y_test = x_test*m.value + b.value

plt.plot(x_test, y_test)
plt.plot(x_train, y_train, '.')
```
